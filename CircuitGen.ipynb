```python
from collections import Counter
from pysat.solvers import Solver
from pysat.formula import CNF
import copy
import pprint

def circGen(ands, inputs, ors, not_composition):
    import random
    ands += 1
    theblock = [[]for i in range(ands)]
    iterator = 0
    num_of_nots = 0
    entries = 0
    
    for i in range(ands + ors):
        while(1):
            if i%inputs + 1 not in theblock[(i+iterator)%ands]:
                theblock[(i+iterator)%ands].append((i%inputs) +1 )
                iterator = 0
                break
            elif( iterator > ands):
                #print("there's a bit extra\n")
                iterator = 0
                break
            else:
                iterator += 1
    
    #Negative assignment part          
    for x in theblock:
        entries = entries + len(x)
    num_of_nots = int(entries*not_composition)
    #print("number of nots is ", num_of_nots)
    
    repeat_cond = True
    
    neg_out_iter = len(theblock)-1 #changes position with each iteration 
    neg_in_iter = len(theblock[neg_out_iter])-1 #this is to decide inside which bracket the not is going to be applied
    thecount = 0
    
    while(repeat_cond == True):
        temp_block = copy.deepcopy(theblock)
        for i in range(num_of_nots):
            while(1):
                if(temp_block[neg_out_iter][neg_in_iter]):
                    #print("Entered outer loop")
                    if(temp_block[neg_out_iter][neg_in_iter] > 0):
                        temp_block[neg_out_iter][neg_in_iter] *= -1 
                        #print("This is the original block"+ str(theblock))
                        #print(temp_block)
                        break
                neg_out_iter = random.randint(0, len(temp_block))-1 #(neg_out_iter-1)%len(temp_block)
                neg_in_iter = random.randint(0, len(temp_block[neg_out_iter]))%len(temp_block[neg_out_iter])
                
        counted = Counter(tuple(sorted(entry)) for entry in temp_block)
        #pprint.pprint(dict(counted))
        repeat_cond = False
        
        #checking if it's satisfiable
        cnf = CNF(from_clauses=temp_block)
        #print(cnf.clauses)
        s = Solver(bootstrap_with=cnf)
        if (s.solve() == False):
            repeat_cond = True
        #print(s.get_model())
        
        #checking for repeated elements
        for element in counted:
            if(counted[element]>1 and thecount < 100):
                repeat_cond = True
                #print("REPEATED ELEMENT")
            if(counted[element]>2 and thecount > 101):
                repeat_cond = True
                #print("TWO REPEATED ELEMENTS")
                
        if(repeat_cond == False):
            theblock = temp_block
        thecount +=1
            
    return theblock
    
print(circGen(3, 3,8, 0.3))
```
